```python
import dask.dataframe as dd
```

```python
df1 = dd.read_csv('itineraries.csv', dtype={'totalTravelDistance': 'float64'})
df1.head()
```

```python
df1 = df1.dropna()
```

```python
df = df1[df1['destinationAirport'] == 'JFK']
df.columns
```

```python
df = df.drop(columns=['legId', 'destinationAirport', 'fareBasisCode', 'segmentsDepartureTimeEpochSeconds',
       'segmentsArrivalTimeEpochSeconds', 'segmentsArrivalAirportCode', 'segmentsDepartureAirportCode', 'segmentsAirlineCode', 'segmentsDurationInSeconds',
       'segmentsDistance'])
df.head()
```

```python
df['segmentsAirlineName'] = df['segmentsAirlineName'].str.split('\|\|').str[0]
df['segmentsCabinCode'] = df['segmentsCabinCode'].str.split('\|\|').str[0]
df['segmentsArrivalTimeRaw'] = df['segmentsArrivalTimeRaw'].str.split('\|\|').str[0]
df['segmentsDepartureTimeRaw'] = df['segmentsDepartureTimeRaw'].str.split('\|\|').str[0]
df['segmentsEquipmentDescription'] = df['segmentsEquipmentDescription'].str.split('\|\|').str[0]
```

```python
df['segmentsArrivalTimeRaw'] = dd.to_datetime(df['segmentsArrivalTimeRaw'], format='%Y-%m-%dT%H:%M:%S.%f%z')
df['segmentsDepartureTimeRaw'] = dd.to_datetime(df['segmentsDepartureTimeRaw'], format='%Y-%m-%dT%H:%M:%S.%f%z')
```

```python
# Assuming df is your Dask DataFrame
df['travelDuration'] = dd.to_timedelta(df['travelDuration'])

# Extract hours and minutes
df['Hour'] = df['travelDuration'].dt.components['hours']
df['Minute'] = df['travelDuration'].dt.components['minutes']
df['searchDate'] = dd.to_datetime(df['searchDate'])
df['flightDate'] = dd.to_datetime(df['flightDate'])
```

```python
# Extract month from 'searchDate' and 'flightDate'
df['searchMonth'] = df['searchDate'].dt.month_name()
df['flightMonth'] = df['flightDate'].dt.month_name()

# If needed, convert Hour and Minute to integers
df['Hour'] = df['Hour'].astype(int)
df['Minute'] = df['Minute'].astype(int)

bool_columns = ['isBasicEconomy', 'isRefundable', 'isNonStop']

# Convert boolean columns to int (0 for False, 1 for True)
df[bool_columns] = df[bool_columns].astype(int)
```

```python
#Extract day from searchDate amd flightDate
df['searchDay'] = df['searchDate'].dt.strftime('%A')
df['flighthDay'] = df['flightDate'].dt.strftime('%A')
df.head()
```

```python
df['searchDate'] = dd.to_datetime(df['searchDate'])
df['flightDate'] = dd.to_datetime(df['flightDate'])

df['dayDiff'] = (df['flightDate'] - df['searchDate']).dt.days
```

```python
# Drop the original 'travelDuration' column if not needed
df = df.drop(columns=['travelDuration', 'segmentsArrivalTimeRaw', 'segmentsDepartureTimeRaw'])
```

```python
df['totalDuration'] = df['Hour']*60 + df['Minute']
```

```python
df.dtypes
```

```python
df = df.reset_index(drop=True)
df.head()
```

```python
df = df.sample(frac=0.1, random_state=42)
```

```python
df.to_csv('jfkDatafinal.csv', index=False, single_file=True)
```
